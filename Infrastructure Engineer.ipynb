```python
# Converting parking_sensors.csv to parking_sensors.parquet

# Define paths
csv_path = "/Volumes/Infrastructure_Engineer_Catalog/Infrastructure_Engineer_Schema/Infrastructure_Engineer_Volume/parking_sensors.csv"

parquet_path = "/Volumes/Infrastructure_Engineer_Catalog/Infrastructure_Engineer_Schema/Infrastructure_Engineer_Volume/parking_sensors.parquet"

# Read CSV and write as Parquet
(
    spark.read
         .option("header", "true")
         .option("inferSchema", "true")
         .csv(csv_path)
         .write
         .mode("overwrite")
         .parquet(parquet_path)
)
```

```python
# After conversion Verify the Parquet File

spark.read.parquet(parquet_path).display()
```

```python
# TASK 1: Schema Enrichment & Joins
# STEP 1: Read Both Datasets from Your Volume

# Paths
sensor_path = "/Volumes/Infrastructure_Engineer_Catalog/Infrastructure_Engineer_Schema/Infrastructure_Engineer_Volume/parking_sensors.parquet"

tariff_path = "/Volumes/Infrastructure_Engineer_Catalog/Infrastructure_Engineer_Schema/Infrastructure_Engineer_Volume/tariff_metadata.csv"

# Read sensor data (Parquet)
sensor_df = spark.read.parquet(sensor_path)

# Read tariff metadata (CSV)
tariff_df = (
    spark.read
         .option("header", "true")
         .option("inferSchema", "true")
         .csv(tariff_path)
)

display(sensor_df)
display(tariff_df)
```

```python
# STEP 2: Perform Schema Enrichment Join
#Key Rules Implemented

#LEFT JOIN (do not lose sensor data)
#Default rate = 2.00
#Flag mismatched records

#PySpark Join Logic

from pyspark.sql.functions import col, when, lit

enriched_df = (
    sensor_df
    .join(tariff_df, on="zone_id", how="left")
    .withColumn(
        "price_per_hour",
        when(col("price_per_hour").isNull(), lit(2.00))
        .otherwise(col("price_per_hour"))
    )
    .withColumn(
        "match_status",
        when(col("neighborhood").isNull(), lit("Unmatched"))
        .otherwise(lit("Matched"))
    )
)
```

```python
# STEP 3: Validate Mismatched Keys
# Only mismatched zones
enriched_df.filter(col("match_status") == "Unmatched").display()
```

```python
# STEP 4 (Recommended): Persist the Enriched Data
# Save as a Managed Delta Table

enriched_df.write \
    .mode("overwrite") \
    .format("delta") \
    .saveAsTable(
        "Infrastructure_Engineer_Catalog.Infrastructure_Engineer_Schema.parking_sensors_enriched"
    )
```

```python
%sql
--TASK 2: Incremental Loading Logic
--STEP 1: Create the Final Target Table (One-Time Setup)

--This table will hold only the latest version of each parking event.
CREATE TABLE IF NOT EXISTS Infrastructure_Engineer_Catalog.Infrastructure_Engineer_Schema.parking_sensors_final
USING DELTA
AS
SELECT *
FROM Infrastructure_Engineer_Catalog.Infrastructure_Engineer_Schema.parking_sensors_enriched
WHERE 1 = 0;
```

```python
# STEP 2: Get Last Processed Timestamp
# We find how far the pipeline has already processed.

from pyspark.sql.functions import max

final_table = "Infrastructure_Engineer_Catalog.Infrastructure_Engineer_Schema.parking_sensors_final"

# Get latest last_modified timestamp
last_processed_ts = (
    spark.table(final_table)
         .select(max("last_modified").alias("max_ts"))
         .collect()[0]["max_ts"]
)

print("Last processed timestamp:", last_processed_ts)
```

```python
# STEP 3: Filter Only New or Updated Records
# This is the core incremental logic.

source_df = spark.table(
    "Infrastructure_Engineer_Catalog.Infrastructure_Engineer_Schema.parking_sensors_enriched"
)

incremental_df = source_df.filter(
    source_df.last_modified > last_processed_ts
)

display(incremental_df)
```

```python
%sql
--STEP 4: Perform UPSERT (MERGE INTO)
--SQL MERGE Logic

MERGE INTO Infrastructure_Engineer_Catalog.Infrastructure_Engineer_Schema.parking_sensors_final AS target
USING (
    SELECT *
    FROM Infrastructure_Engineer_Catalog.Infrastructure_Engineer_Schema.parking_sensors_enriched
    WHERE last_modified >
          (SELECT COALESCE(MAX(last_modified), '1900-01-01')
           FROM Infrastructure_Engineer_Catalog.Infrastructure_Engineer_Schema.parking_sensors_final)
) AS source
ON target.spot_id = source.spot_id
AND target.event_time = source.event_time

WHEN MATCHED THEN
  UPDATE SET *

WHEN NOT MATCHED THEN
  INSERT *;
```
